```python
import pandas as pd
import numpy as np

file_to_load = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)
```

```python
# Display Player count
player_info= purchase_data.groupby(["SN"])
total_players = len(player_info)
#print(total_players)
player_ct_table = pd.DataFrame({"Total Players":total_players}, index =[0])
print(player_ct_table)
```

```python
##Purchase Analysis
items= purchase_data.groupby(["Item ID","Item Name"])
#print(items)
items_unique =len(items)
#print(items_unique)
avg_price= purchase_data["Price"].mean()
total_purchase =len(purchase_data)
total_revenue =purchase_data["Price"].sum()
#print(total_revenue)
```

```python
purchase_summary=pd.DataFrame({"Number of Unique Items":items_unique, "Average Price":avg_price,"Number of Purchases":total_purchase,"Total Revenue":total_revenue},index=[0])
purchase_summary= purchase_summary[["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"]]
pd.options.display.float_format = '{:,.2f}'.format
purchase_summary
```

```python
##Gender Demographics
#find unique players & sort by gender-find the percentage-make table and display.
players_unique= purchase_data.drop_duplicates(subset="SN",keep="first")
gender_ct =players_unique["Gender"].value_counts()
#gender_ct.head()
player_percent=(gender_ct/total_players)*100
#print(player_percent)

#styling - 
gender_ct= pd.DataFrame(gender_ct)
gender_ct= gender_ct.rename(columns={"Gender": "Total Count"})
gender_ct["Percentage of Players"]= player_percent.values
gender_ct =gender_ct[["Percentage of Players","Total Count"]]
pd.options.display.float_format = '{:,.2f}'.format
gender_ct
```

```python
##Gender Purchase analysis
#sort out gender from the main purchse file and creat a datafram-create avg price and total purchse by gender-merge dfs
gender_cts= purchase_data["Gender"].value_counts()
gender_cts= pd.DataFrame(gender_cts)

gender_avg= purchase_data.groupby("Gender").mean()
gender_avg= gender_avg[["Price"]]
gender_avg= gender_avg.rename(columns={"Price": "Average Purchase Price"})
gender_combined=pd.concat([gender_avg,gender_ct],axis=1)
gender_combined= gender_combined.rename(columns={"Gender":"Purchase Count"})

gender_purchase= purchase_data.groupby("Gender").sum()
gender_purchase= gender_purchase[["Price"]]
gender_purchase =gender_purchase.rename(columns={"Price": "Total Purchase Value"})
gender_purchases=pd.concat([gender_combined, gender_purchase],axis=1)
gender_purchases

avg_total_per_person= gender_purchase["Total Purchase Value"] / gender_ct["Total Count"]
#avg_total_per_person= avg_total_per_person.rename(columns={""})

gender_total_table= pd.concat([gender_purchases, avg_total_per_person], axis=1)
gender_total_table =gender_total_table.rename(columns={"0": "Average Per Person"})
pd.options.display.float_format = '{:,.2f}'.format
gender_total_table
```

```python
#Age Demographics
#make bins-
bins = [0, 9, 14, 19, 24, 29, 34, 39, 99]
bin_names = ["<10", "10-14", "15-19" ,"20-24","25-29","30-34", "35-39","40+"]

players_unique["Age Group"] = pd.cut(players_unique["Age"],bins,labels=bin_names)
age_demo= players_unique.groupby("Age Group").count()
age_demo= age_demo[["Age"]]
age_demo = age_demo.rename(columns={"Age": "Total Count"})
age_demo["Percentage of Players"] = age_demo["Total Count"] / total_players * 100
age_demo = age_demo[["Percentage of Players", "Total Count"]]
pd.options.display.float_format = '{:,.2f}%'.format
age_demo
```

```python
##Age Purchase Analysis
#sort out gender from the main purchse file and creat a datafram-create avg price and total purchse by age-merge dfs


purchase_data["Age Group"] = pd.cut(purchase_data["Age"],bins,labels=bin_names)
age_grp = purchase_data.groupby("Age Group").count()
age_grp= age_grp[["Age"]]
#age_grp

age_grp= age_grp.rename(columns={"Age": "Purchase Count"})
age_avg = purchase_data.groupby("Age Group").mean()
age_avg= age_avg[["Price"]]
age_avg= age_avg.rename(columns={"Price": "Average Purchase Price"})
age_combined= pd.concat([age_avg, age_grp], axis=1)

age_total= purchase_data.groupby("Age Group").sum()
age_total= age_total[["Price"]]
age_total= age_total.rename(columns={"Price": "Total Purchase Value"})
age_purchase= pd.concat([age_combined, age_total], axis=1)

avg_total_per_person= age_purchase["Total Purchase Value"] / age_demo["Total Count"]
age_total_table= pd.concat([age_purchase, avg_total_per_person], axis=1)
age_total_table = age_total_table.rename(columns={"Price": "Average Purchase Price"})

#age_table = age_total_table[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Average Per Person"]]
pd.options.display.float_format = "${:,.2f}".format

age_total_table
```

```python
#Popular Items by spending
popular_items= items[["Price"]].count().nlargest(5,"Price")
popular_items = popular_items.rename(columns={"Price": "Purchase Count"})

item_price= items[["Price"]].mean()
item_price= item_price.rename(columns={"Price": "Item Price"})
item_combined =popular_items.join(item_price)

item_purchase =items[["Price"]].sum()
item_purchase= item_purchase.rename(columns={"Price": "Total Purchase Value"})

top_items =item_combined.join(item_purchase)

pd.options.display.float_format= "${:,.2f}".format
top_items
```

```python
##Top Profitable Items-
item_sales= items[["Price"]].sum().nlargest(5,"Price")
item_sales =item_sales.rename(columns={"Price": "Total Purchase Value"})
item_sales

item_purchase =items[["Price"]].count()
item_purchase= item_purchase.rename(columns={"Price": "Purchase Count"})
profit_combined= item_sales.join(item_purchase)

item_price = items[["Price"]].mean()
item_price= item_price.rename(columns={"Price": "Item Price"})
profit_table= profit_combined.join(item_price)

profit_item_sales= profit_table[["Purchase Count", "Item Price", "Total Purchase Value"]]
pd.options.display.float_format ='${:,.2f}'.format

profit_item_sales
```
